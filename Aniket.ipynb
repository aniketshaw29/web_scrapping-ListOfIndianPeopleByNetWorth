```python
import requests
link="https://en.m.wikipedia.org/wiki/List_of_Indian_people_by_net_worth"
link_text=requests.get(link).text
print(link_text)
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(link_text,'lxml')
print(soup)
```

```python
print(soup.prettify())
```

```python
print(soup.title)
```

```python
print(soup.title.string)
```

```python
soup.find_all('a')
```

```python
all_table=soup.find_all('table')
```

```python
print(all_table)
```

```python
our_table=soup.find('table',class_='wikitable sortable')
```

```python
print(our_table)
```

```python
table_links=our_table.find_all('a')
```

```python
print(table_links)
```

```python
bill=[]
for links in table_links:
    bill.append(links.get('title'))
print(bill)
```
